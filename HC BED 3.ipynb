```python
def cp_ig_coeff_pseudo(kw , w): #T IN KELVIN  #joule/mole
    a0 = -1.41779 + 0.11828 * kw
    a1 = -1 * (6.99724 - 8.69326*kw + 0.27715*kw**2) * 10**(-4)
    a2 = -2.2582 * 10**(-6)

    b0 = 1.09223 - 2.48245*w
    b1 = -1 * (3.434 - 7.14*w) * 10**(-3)
    b2 = -1 * (7.2661 - 9.2561*w) * 10**(-7)

    c = ( ( (12.8-kw)*(10-kw) ) / (10*w) )**2

    return a0 , a1 , a2 , b0 , b1 , b2 , c

def cp_liq_coeff_pseudo(kw , sg ): #kj/kg.k ----> convert it to kmole then multiply by mwt ---> kj/kmole.k
   
   a1 = -4.90383 + kw*(0.099319 + 0.104281*sg) + ((4.81407 - 0.194833*kw)/sg)
   a2 = (7.53621 + 6.21461 * kw) * (1.12172 - (0.27634/sg)) * 10**(-4)
   a3 = -1*(1.35652 + 1.11863*kw)* (2.9027 - (0.70958/sg)) * 10**(-7)
   
   
   return  a1 , a2 , a3
```

```python
def Quench(FF_vap_ , FD_vap_ , FK_vap_ , FN_vap_ , FG_vap_ , FF_liq_ , FD_liq_ , FN_liq_ , FG_liq_ , FK_liq_, F_H2_,T_ref = 368.4 , Ti = 370.5056866345765):

    #MWTs
    MWT_VGO = 447.8479509
    MWT_N   = 114.9170391
    MWT_D   = 292.7362979
    MWT_K   = 170.23913
    MWT_G   = 16.56482354
    MWT_R   = 759.9991037 
    #CALC T_avg
    T = 0.5 * (T_ref+273.15 + Ti+273.15) 
    
    #flow rates
    #FF_vap_ = 15.840102483329341
    #FD_vap_   = 4.463853976824296
    #Fk_vap_   = 6.545354265675065
    #FN_vap_   = 3.4293178177902788
    #FG_vap_   = 81.38578465171908
    #FF_liq_   = 52.654321656969614
    #fd_liq   = 101.08297105104931
    #fn_liq   = 27.567904370967565
    #fg_liq   = 7.764397700601099
    #fk_liq   = 70.57487224343332
    f_r_liq  = 4.745247617
    f_r_vap  = 0.005588344

    #F_H2      =3259.362083176698
    F_H2S_liq = 18.06256176
    F_H2S_vap = 297.7673451
    F_NH3_vap = 2.52506041
    F_NH3_liq = 0.149586238
    #calc cp_avg for each and every component
    #note didn't crack the lumps into the old components
    a0_VGO_vap , a1_VGO_vap , a2_VGO_vap , b0_VGO_vap , b1_VGO_vap  , b2_VGO_vap , c_VGO_vap = cp_ig_coeff_pseudo(kw = 11.83301505, w = 1.046734155)
    a0_D_vap , a1_D_vap , a2_D_vap , b0_D_vap , b1_D_vap , b2_D_vap , c_D_vap = cp_ig_coeff_pseudo(kw = 11.82860779, w = 0.77914103)
    a0_K_vap , a1_K_vap , a2_K_vap , b0_K_vap , b1_K_vap , b2_K_vap , c_K_vap = cp_ig_coeff_pseudo(kw = 11.93827339, w = 0.518610817)
    a0_N_vap , a1_N_vap , a2_N_vap , b0_N_vap , b1_N_vap , b2_N_vap , c_N_vap = cp_ig_coeff_pseudo(kw = 12.20555753, w = 0.345207156)
    #a0_G_vap , a1_G_vap , a2_G_vap , b0_G_vap , b1_G_vap , b2_G_vap , c_G_vap =  #cp_ig_coeff_pseudo(kw = 13.3, w = 1.47E-02)  #here's the problem
    a0_R_vap , a1_R_vap , a2_R_vap , b0_R_vap , b1_R_vap , b2_R_vap , c_R_vap = cp_ig_coeff_pseudo(kw = 11.81969833, w = 1.558194126)

    a1_VGO_liq , a2_VGO_liq , a3_VGO_liq  = cp_liq_coeff_pseudo(kw = 11.833263  , sg = 0.922061456)  
    a1_D_liq   , a2_D_liq   , a3_D_liq    = cp_liq_coeff_pseudo(kw = 11.82860779, sg = 0.872109193)
    a1_K_liq   , a2_K_liq   , a3_K_liq    = cp_liq_coeff_pseudo(kw = 11.93827339, sg = 0.803929666)
    a1_N_liq   , a2_N_liq   , a3_N_liq    = cp_liq_coeff_pseudo(kw = 12.20555753, sg = 0.734974159)
    a1_G_liq   , a2_G_liq   , a3_G_liq    = cp_liq_coeff_pseudo(kw = 13.3    , sg = 0.566285714)
    a1_R_liq   , a2_R_liq   , a3_R_liq    = cp_liq_coeff_pseudo(kw = 11.81969833, sg = 0.991019934)

    #F * Cp_avg KMOLE/HR  * KJ/KMOLE .K --> KJ/HR
    cp_H2      = F_H2_     * (25.399 + 2.0178e-2 * T  - 3.8549e-5 * T**2 + 3.188e-8 * T**3 - 8.7585e-12 * T**4)
    #print(cp_H2)
    cp_H2S_vap =F_H2S_vap * (34.5234 - 0.01765 * T + 6.77e-5 * T**2 -5.3e-8 * T**3 +1.41e-11 * T**4)
    #print(cp_H2S_vap)
    cp_NH3_vap =F_NH3_vap* (27.55   - 0.025628* T + 9.9e-6   *T**2 -6.7e-9      * T**3)
    #print(cp_NH3_vap)
    cp_NH3_liq =F_NH3_liq* (20.1494 + 0.845765* T -0.00406745*T**2 +6.60687E-06 * T**3)
    #print(cp_NH3_liq)
    cp_H2S_liq =F_H2S_liq * (21.8238 + 0.774223* T -0.00420204*T**2 +7.38677E-06 * T**3)
    #print(cp_H2S_liq)
    cp_F_liq   =FF_liq_* MWT_VGO * (a1_VGO_liq + a2_VGO_liq*T + a3_VGO_liq*T**2)  
    #print(cp_F_liq)
    cp_F_vap   =FF_vap_ *MWT_VGO * (a0_VGO_vap + a1_VGO_vap*T + a2_VGO_vap*T**2 - c_VGO_vap*(b0_VGO_vap + b1_VGO_vap*T +b2_VGO_vap*T**2 ) )
    #print(cp_F_vap)
    cp_D_liq   =FD_liq_* MWT_D   * (a1_D_liq   + a2_D_liq*T   + a3_D_liq    *T**2)     
    #print(cp_D_liq)
    cp_D_vap   =FD_vap_* MWT_D   * (a0_D_vap   + a1_D_vap*T   + a2_D_vap*T**2   - c_D_vap  *(b0_D_vap   + b1_D_vap*T   +b2_D_vap  *T**2   ) )
    #print(cp_D_vap)
    cp_N_liq   =FN_liq_* MWT_N   * (a1_N_liq   + a2_N_liq*T   + a3_N_liq    *T**2)     
    #print(cp_N_liq)
    cp_N_vap   =FN_vap_* MWT_N   * (a0_N_vap   + a1_N_vap*T   + a2_N_vap*T**2   - c_N_vap  *(b0_N_vap   + b1_N_vap*T   +b2_N_vap  *T**2   ) )
    #print(cp_N_vap)
    cp_G_liq   =FG_liq_* MWT_G   * (a1_G_liq   + a2_G_liq*T   + a3_G_liq    *T**2)     
    #print(cp_G_liq)
    cp_G_vap   = 9468.486872 #fg_vap *MWT_G   * (a0_G_vap   + a1_G_vap*T   + a2_G_vap*T**2   - c_G_vap  *(b0_G_vap   + b1_G_vap*T   +b2_G_vap  *T**2   ) )
    #print(cp_G_vap)
    cp_K_liq   =FK_liq_* MWT_K   * (a1_K_liq   + a2_K_liq*T   + a3_K_liq    *T**2)
    #print(cp_K_liq)
    cp_K_vap   =FK_vap_* MWT_K   * (a0_K_vap   + a1_K_vap*T   + a2_K_vap*T**2   - c_K_vap  *(b0_K_vap   + b1_K_vap*T   +b2_K_vap  *T**2   ) )
    #print(cp_K_vap)
    cp_R_liq   =f_r_liq* MWT_R   * (a1_R_liq   + a2_R_liq*T   + a3_R_liq    *T**2) 
    #print(cp_R_liq)
    cp_R_vap   =f_r_vap* MWT_R   * (a0_R_vap   + a1_R_vap*T   + a2_R_vap*T**2   - c_R_vap  *(b0_R_vap   + b1_R_vap*T   +b2_R_vap  *T**2   ) )
    #print(cp_R_vap)

    sum_F_Cp =(cp_H2  +  cp_H2S_vap + cp_NH3_vap + cp_NH3_liq  + cp_H2S_liq + cp_F_liq + cp_F_vap + cp_D_liq + cp_D_vap +  cp_N_liq + cp_N_vap + cp_G_liq + cp_G_vap + cp_K_liq + cp_K_vap)
    delta_T = (Ti - T_ref)

    F_Cp_dt = sum_F_Cp * delta_T
    #print(F_Cp_dt)
    dt_quench = T_ref - 65.56
    T_avg_quench = 0.5 * (T_ref+273.15 + 65.6+273.15) 
    
    cp_dt_quench = (25.399 + 2.0178e-2 * T_avg_quench  - 3.8549e-5 * T_avg_quench**2 + 3.188e-8 * T_avg_quench**3 - 8.7585e-12 * T_avg_quench**4) * dt_quench

    F_H2_quench = F_Cp_dt / cp_dt_quench
    #print(delta_T)
    return F_H2_quench
    
```
